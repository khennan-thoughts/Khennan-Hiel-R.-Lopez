```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"raw\chip-dataset-cleaned.csv",
                 delimiter=",")
df.info()
```

```python
df = df.drop(df.columns[[10,11,12]],axis=1)
df.head()
```

## N Chips Manufactured

```python
df_w = df.copy()
```

```python
chip_count = df_w['Vendor'].value_counts()
chip_count
```

```python
type(chip_count)
```

```python
chart = chip_count.plot(kind='bar',
                        color = ['#FF0A54','#FF477E','#FF5C8A','#FF7096','#FF85A1'],
                        figsize=(12,5))
chart.bar_label(chart.containers[0], label_type='edge')
plt.title('Chips Manufactured (CPU+GPU)',
          fontfamily = 'Verdana',
          fontsize=15,
          fontweight='bold')
plt.xlabel('')
chart.set_xticklabels(chip_count.index, rotation=0)
plt.ylabel('N Chips')
plt.grid(axis='y', linestyle ='-', alpha=1)
plt.show()
```

## N CPU Manufactured

```python
df_cpu = df_w[df_w['Type']=='CPU']
df_cpu
```

```python
cpu_count = df_cpu['Vendor'].value_counts()
cpu_count
```

```python
chart = cpu_count.plot(kind='bar',
                       color = ['#FF0A54','#FF477E'],
                       figsize=(12,5))
plt.title('CPU Manufactured',
           fontfamily = 'Verdana',
           fontsize=15,
           fontweight='bold')
plt.ylabel('N Count')
plt.xlabel('')
chart.set_xticklabels(cpu_count.index, rotation=0)
plt.grid(axis='y', linestyle ='-', alpha=1)
plt.show()
```

## Lowest and Highest TDP (Thermal Design Power)

```python
df_tdp = df_w[['Product','TDP']]
df_tdp
```

```python
type(df_tdp)
```

```python
df_tdphighest = df_tdp.copy()
df_tdplowest = df_tdp.sort_values(by='TDP')
df_tdplowest
```

```python
df_tdplowest = df_tdplowest.head(10)
df_tdplowest
```

```python
plt.figure(figsize=(12,5))
plt.bar(df_tdplowest['Product'],df_tdplowest['TDP'],
        color = ['#FF0A54','#FF477E','#FF5C8A','#FF7096','#FF85A1','#FF99AC','#FBB1BD','#F9BEC7','#F7CAD0','#FAE0E5'])
plt.title('Products with the Lowest TDP',
          fontfamily = 'Verdana',
          fontsize=15,
          fontweight='bold')
plt.ylabel('Watts')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle ='-', alpha=1)
plt.show()
```

```python
df_tdphighest = df_tdphighest.sort_values(by='TDP', ascending=False)
df_tdphighest
```

```python
df_tdphighest = df_tdphighest.head(10)
df_tdphighest
```

```python
plt.figure(figsize=(12,5))
plt.bar(df_tdphighest['Product'],df_tdphighest['TDP'],
        color = ['#FF0A54','#FF477E','#FF5C8A','#FF7096','#FF85A1','#FF99AC','#FBB1BD','#F9BEC7','#F7CAD0','#FAE0E5'])
plt.title('Products with the Highest TDP',
          fontfamily = 'Verdana',
          fontsize=15,
          fontweight='bold')
plt.ylabel('Watts')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle ='-', alpha=1)
plt.show()
```

## GPU Process Size Trend

```python
df_ps = df_w[['Type','Date','Process_Size']].where(df_w['Type']=='GPU').dropna()
df_ps
```

```python
df_ps['Date'] = pd.to_datetime(df_ps['Date'])
df_ps['Year'] = df_ps['Date'].dt.year
df_ps 
```

```python
df_ps = df_ps.groupby('Year')['Process_Size'].mean().reset_index()
df_ps
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_ps['Year'],df_ps['Process_Size'],
         color = '#FF0A54')
plt.title('GPU Process Size Trend',
          fontfamily = 'Verdana',
          fontsize=15,
          fontweight='bold')
plt.ylabel('nm')
plt.grid(axis='y', linestyle ='-', alpha=1)
plt.show()
```
