## Data Cleaning - Replacing Nan with Mean

```python
df_l.info()
```

```python
df_l.isna().sum()
```

```python
def convert_height(h):
    if isinstance(h, str) and '-' in h:
        try:
            feet, inches = map(int, h.split('-'))
            return (feet * 12) + inches
        except ValueError:
            return None
    return h

df_l['Height'] = df_l['Height'].apply(convert_height)

df_l['Height'] = pd.to_numeric(df_l['Height'], errors='coerce')
df_l['Height'] = df_l['Height'].fillna(df_l['Height'].mean())

print(df_l.isna().sum())
```

```python
n_columns = ['Weight',
            '40yd', 
            'Vertical',
            'Bench',
            'Broad Jump',
            '3Cone',
            'Shuttle']

df_l[n_columns] = df_l[n_columns].apply(lambda n: n.fillna(n.mean()))
print(df_l.isna().sum())
```

```python
df_l.info()
```

### Pick

```python

dfcleaned = pd.read_csv("raw/nfl_combine_2010_to_2023_cleaned.csv", sep=",")
dfcleaned['Pick'] = pd.to_numeric(dfcleaned['Pick'], errors='coerce')


if 'Year' in dfcleaned.columns and 'Pick' in dfcleaned.columns:
    highest_picks_per_year = dfcleaned.groupby('Year', as_index=False)['Pick'].max()
    lowest_picks_per_year = dfcleaned.groupby('Year', as_index=False)['Pick'].min()

    highest_picks_per_year.rename(columns={'Pick': 'Highest Pick'}, inplace=True)
    lowest_picks_per_year.rename(columns={'Pick': 'Lowest Pick'}, inplace=True)

    picks_summary = highest_picks_per_year.merge(lowest_picks_per_year, on='Year')

    plt.figure(figsize=(12, 6))
    plt.plot(picks_summary['Year'], picks_summary['Highest Pick'], marker='o', color='#FF477E', label='Highest Pick', linestyle='-', linewidth=2)
    plt.plot(picks_summary['Year'], picks_summary['Lowest Pick'], marker='o', color='#E91E63', label='Lowest Pick', linestyle='-', linewidth=2)

    for i in range(len(picks_summary)):
        plt.text(picks_summary['Year'][i], picks_summary['Highest Pick'][i], str(picks_summary['Highest Pick'][i]), 
                 ha='center', va='bottom', fontsize=10, fontweight='bold', color='#FF477E')
        plt.text(picks_summary['Year'][i], picks_summary['Lowest Pick'][i], str(picks_summary['Lowest Pick'][i]), 
                 ha='center', va='top', fontsize=10, fontweight='bold', color='#E91E63')

    plt.xlabel('Year', fontfamily='Verdana', fontsize=12, fontweight='bold')
    plt.ylabel('Pick Number', fontfamily='Verdana', fontsize=12, fontweight='bold')
    plt.title('Highest & Lowest Picks Per Year (2010-2023)', fontfamily='Verdana', fontsize=14, fontweight='bold')

    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend()
    plt.xticks(rotation=0)

    plt.show()

else:
    print("Error: Required columns not found.")
```
